```python
# Handle Missing values 

def handle_missing_values(df, feature):
    if df[feature].dtype == 'object':

        # Missing indicator for categorical feature

        df[f"{feature}_missing_ind"] = df[feature].isnull().astype(int)

        df[feature] = df[feature].fillna("missing")
    else:
        df[feature] = df[feature].fillna(df[feature].mean())
            
    return df
```

```python
missing_feature = [feature for feature in df.columns if df[feature].isnull().mean() > 0]
```

```python
# Drop job_id

df = df.drop('job_id', axis=1)
```

```python
import string
import re
```

```python
# Text Cleaning - removing unwanted characters and normalize text

def text_cleaning(text):

    text = text.lower()

    text = re.sub(r"['`]", "", text)

    # text = text.translate(str.maketrans("", "", string.punctuation))

    text = re.sub(f"[{re.escape(string.punctuation)}]", " ", text)

    # Remove extra whitespaces

    text = " ".join(text.split())

    return text

```

```python
cat_features = [feature for feature in df.columns if df[feature].dtype not in [int, float]]
num_features = [feature for feature in df.columns if df[feature].dtype in [int, float]]
```

```python
# Apply text cleaning to all the features

for feature in cat_features:
        
    df[f"{feature}_cleaned"] = df[feature].apply(lambda x: text_cleaning(x))
```

```python
# Lemmatization 

import spacy 

from tqdm import tqdm

nlp = spacy.load("en_core_web_sm")
```

```python
text = filtered_df["text_all"].tolist()

lemmatized_text = []

for doc in tqdm(nlp.pipe(text, batch_size=100, disable=["parser", "ner"]), total=len(text)):

    lemmas = [token.lemma_ for token in doc if not token.is_punct and not token.is_space]
    lemmatized_text.append(" ".join(lemmas)) 
```

```python
lemmatized_text
```

```python
filtered_df["text_lemma"] = lemmatized_text
```

```python
# Now split the data into train and test dataset
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(filtered_df.drop("fraudulent", axis=1), filtered_df["fraudulent"], test_size=0.2, random_state=42, stratify=filtered_df["fraudulent"])
```

```python
# Apply One Hot Encoding 

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore')

X_train_encoded = encoder.fit_transform(X_train[cat_features])
X_test_encoded = encoder.transform(X_test[cat_features])
```

```python
X_train_encoded
```

```python
X_train[num_features]
```

```python
# Now Apply TF-ID vectorizer on text_lemma

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))

X_train_text = tfidf.fit_transform(X_train["text_lemma"])
X_test_text = tfidf.transform(X_test["text_lemma"])
```

```python
X_train_num = X_train[num_features]
X_test_num = X_test[num_features]

X_train_num = np.array(X_train_num)
X_test_num = np.array(X_test_num)
```

```python
X_train_encoded.shape
```

```python
from scipy.sparse import hstack

X_train_final = hstack([X_train_num, X_train_encoded, X_train_text])
X_test_final = hstack([X_test_num, X_test_encoded, X_test_text])
```

```python
print(X_train_final.shape)
print(type(X_train_final))
```

```python
print(filtered_df["fraudulent"].value_counts(normalize=True))
```

```python
# Handle Imbalanced data

from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_train_resampled, y_train_resampled = smote.fit_resample(X_train_final, y_train)
```

```python
# Simple testing on model after oversampling

from sklearn.linear_model import LogisticRegression

logistic = LogisticRegression()
```

```python
logistic.fit(X_train_resampled, y_train_resampled)
```

```python
y_pred_logistic = logistic.predict(X_test_final)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, classification_report, f1_score
```

```python
acc_score_log = accuracy_score(y_test, y_pred_logistic)
precision_score_log = precision_score(y_test, y_pred_logistic)
recall_score_log = recall_score(y_test, y_pred_logistic)
f1_score_log = f1_score(y_test, y_pred_logistic)
roc_log = roc_auc_score(y_test, y_pred_logistic)
cr_log = classification_report(y_test, y_pred_logistic)
```

```python
print(f"Accuracy Score Log: {acc_score_log:.2f}%")
print(f"Precision Score Log: {precision_score_log:2f}%")
print(f"Recall Score Log: {recall_score_log:.2f}%")
print(f"F1 Score Log: {f1_score_log:.2f}%")
print(f"ROC AUC SCORE log: {roc_log:.2f}%")
print(f"\nClassification Report Log: \n{cr_log}")
```

### Summary of Feature engineering notebook

In this notebook I did the feature engineering from start to the end of NLP - __Fake Job Prediction__

Basic flow of feature engineering of any NLP is that 

- Handle Missing value and always give missing indicator feature also
- Clean the text (remove punctuation, remove whitespace)
- Combine all rich text features into one feature like in this `text_all`
- remove stop words
- detect language - and if english was below 80% then translate otherwise leave it
- Correct spelling mistake and typos (optional)
- Apply lemmatization
- Split into train and test dataset
- One Hot encoding on normal features 
- Apply Text Vectorizer on Rich text feature like in this `text_all`
- And combine the data together using `scipy.sparse.hstack` because when we have more dimensions than normal then we use sparse matrix.
- If you are dataset is imbalanced then try to Oversample the data using `SMOTE`.
