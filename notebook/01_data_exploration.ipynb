```python
def missing_handles(df, feature_name):

    """
    Missing handles function handle and fill missing values and make missing indicator also
    """

    # Missing value indicator
    df[f"missing_indicator_{feature_name}"] = df[feature_name].isnull().astype(int)

    # Fill missing values with "missing" placeholder

    df[feature_name] = df[feature_name].fillna("missing")
```

```python
for feature in missing_cols:

    missing_handles(eda_df, feature)
```

```python
# Term Frequency Analysis

import re
import string
```

```python
def text_cleaning(text):

    # Convert text to lowercase
    text = text.lower()

    # Remove punctuation
    translate = str.maketrans('', '', string.punctuation)
    text = text.translate(translate)

    return text
```

```python
eda_df["description_cleaned"] = eda_df["description"].map(lambda x: text_cleaning(x))

eda_df["company_profile_cleaned"] = eda_df["company_profile"].map(lambda x: text_cleaning(x))

eda_df["requirment_cleaned"] = eda_df["requirements"].map(lambda x: text_cleaning(x))

eda_df["benefits_cleaned"] = eda_df["benefits"].map(lambda x: text_cleaning(x))
```

```python
from nltk.corpus import stopwords
from collections import Counter
import nltk

nltk.download("stopwords")

nltk.download("punkt_tab")
```

```python
stop_words = set(stopwords.words('english'))
```

```python
description_text_data = " ".join(eda_df["description_cleaned"].astype(str))
```

```python
description_tokens = nltk.word_tokenize(description_text_data)
```

```python
word_freq = Counter(description_tokens)
```

```python
word_freq.most_common(10)
```

```python
filtered_word = [word for word in description_tokens if word not in stop_words]

filtered_word = Counter(filtered_word)

filtered_word.most_common(10)
```

```python
common_words = filtered_word.most_common(10)

word, count = zip(*common_words)

plt.figure(figsize=(10, 6))

sns.barplot(x=word, y=count)

plt.title("Most Common Frequent Word in Description")

plt.show()
```

```python
fake_job = eda_df[eda_df["fraudulent"] == 1]["description_cleaned"].astype(str)
real_job = eda_df[eda_df["fraudulent"] == 0]["description_cleaned"].astype(str)
```

```python
fake_job.head()
```

```python
fake_text = " ".join(fake_job)
real_text = " ".join(real_job)
```

```python
fake_text = text_cleaning(fake_text)
real_text = text_cleaning(real_text)
```

```python
fake_token = nltk.word_tokenize(fake_text)
real_token = nltk.word_tokenize(real_text)
```

```python
fake_token = [word for word in fake_token if word not in stop_words]
real_token = [word for word in real_token if word not in stop_words]
```

```python
fake_common = Counter(fake_token).most_common(20)
real_common = Counter(real_token).most_common(20)
```

```python
f_word, f_count = zip(*fake_common)
```

```python
plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 1)
sns.barplot(x=f_word, y=f_count)
plt.title("frequent word in Fake Job")
plt.xticks(rotation=90)

plt.subplot(1, 2, 2)
r_word, r_count = zip(*real_common)
sns.barplot(x=r_word, y=r_count)
plt.title("Frequent word in real job")
plt.xticks(rotation=90)
```

```python
# uniquness in fake and real job word

fake_set = set([word for word, _ in fake_common])
real_set = set([word for word, _ in real_common])

unique_fake_words = fake_set - real_set
unique_real_words = real_set - fake_set

print(f"Unique words to fake job: {unique_fake_words}")
print(f"Unique words to real job: {unique_real_words}")
```
