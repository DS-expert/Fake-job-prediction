```python
%run ../notebook/notebook_setup.py
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.sparse import load_npz
from config.config import TRAIN_FEATURES_PATH, TRAIN_TARGET_PATH, TEST_FEATURES_PATH, TEST_TARGET_PATH
```

```python
X_train = load_npz(TRAIN_FEATURES_PATH)
y_train = np.load(TRAIN_TARGET_PATH)
X_test = load_npz(TEST_FEATURES_PATH)
y_test = np.load(TEST_TARGET_PATH)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report

acc_score_lr = accuracy_score(y_test, y_pred_lr)
roc_lr = roc_auc_score(y_test, y_pred_lr)
cr_lr = classification_report(y_test, y_pred_lr)

print(f"Accuracy Score Linear Regression: {acc_score_lr:.2f}%")
print(f"ROC Accuracy Score Linear Regression: {roc_lr:.2f}%")
print(f"Classification report Linear Regression: \n{cr_lr}")
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()

tree.fit(X_train, y_train)
```

```python
y_pred_tree = tree.predict(X_test)
```

```python
acc_score_tree = accuracy_score(y_test, y_pred_tree)
roc_tree = roc_auc_score(y_test, y_pred_tree)
cr_tree = classification_report(y_test, y_pred_tree)

print(f"Accuracy Score Tree: {acc_score_tree:.2f}%")
print(f"ROC Accuracy Score Tree: {roc_tree:.2f}%")
print(f"Classification Report Tree: \n{cr_tree}")
```
