```python
def explore_null_feature(df, feature_name):

    if df[feature_name].isnull().mean() > 0:

        print(f"*******************\n")
        print(f"Feature Name: {feature}")
        print(df[feature_name].value_counts())
        print(f"\n************************\n")
```

```python
for feature in df.columns:

    explore_null_feature(df, feature)
```

```python
for feature in df.columns:

    if df[feature].isnull().mean() > 0:

        missing_ratio = df[feature].isnull().mean() * 100
        print(f"{feature} {missing_ratio:.2f}% missing")

```
